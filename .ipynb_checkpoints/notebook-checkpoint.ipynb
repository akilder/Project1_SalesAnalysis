```python
import pandas as pd
from utils import generate_random_sales

dates = pd.date_range(start='2025-01-01', end='2025-12-01', freq='MS')

sales_a = generate_random_sales(50, 100, 12)
sales_b = generate_random_sales(30, 80, 12)
sales_c = generate_random_sales(20, 60, 12)
sales_d = generate_random_sales(10, 50, 12)

df = pd.DataFrame({
    'Date': dates,
    'Product_A': sales_a,
    'Product_B': sales_b,
    'Product_C': sales_c,
    'Product_D': sales_d
})

df.to_csv('data/initial.csv', index=False)
```

```python
# Add total and average sales
month_names = dates.month_name()

df_M=pandas.Dataframe({
    'Month': month_names,
    'Product_A': sales_a,
    'Product_B': sales_b,
    'Product_C': sales_c,
    'Product_D': sales_d
})
df['Total Sales'] = df[['Product A','Product B','Product C','Product D']].sum(axis=1)
df['Average Sales'] = df[['Product A','Product B','Product C','Product D']].mean(axis=1)

# Month-over-month growth
df['MoM Growth (%)'] = df['Total Sales'].pct_change() * 100

# Assign quarters
def assign_quarter(month):
    if month in [1,2,3]: return 'Q1'
    elif month in [4,5,6]: return 'Q2'
    elif month in [7,8,9]: return 'Q3'
    else: return 'Q4'

df['Quarter'] = df['Date'].dt.month.map(assign_quarter)

# Max and min sales product per month
df['Max_Sales_Product'] = df[['Product A','Product B','Product C','Product D']].idxmax(axis=1)
df['Min_Sales_Product'] = df[['Product A','Product B','Product C','Product D']].idxmin(axis=1)

# Save final.csv
df.to_csv('data/final.csv', index=False)
```

```python
# Average sales per quarter per product
pivot_avg = df.pivot_table(index='Quarter', values=['Product A','Product B','Product C','Product D','Total Sales'], aggfunc='mean')

# Total sales per quarter
pivot_total = df.pivot_table(index='Quarter', values=['Product A','Product B','Product C','Product D','Total Sales'], aggfunc='sum')

# Save pivot tables in Excel
with pd.ExcelWriter('data/output.xlsx') as writer:
    pivot_avg.to_excel(writer, sheet_name='Avg Sales')
    pivot_total.to_excel(writer, sheet_name='Total Sales')
```

```python
best_month = df.loc[df['Total Sales'].idxmax(), 'Date']
best_product = df[['Product A','Product B','Product C','Product D']].sum().idxmax()
best_quarter = df.groupby('Quarter')['Total Sales'].sum().idxmax()

print(f"Best Month: {best_month}")
print(f"Best Product: {best_product}")
print(f"Best Quarter: {best_quarter}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line chart
plt.figure(figsize=(10,5))
for product in ['Product A','Product B','Product C','Product D']:
    plt.plot(df['Date'], df[product], marker='o', label=product)
plt.title('Monthly Sales per Product')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Stacked bar chart
df.set_index('Date')[['Product A','Product B','Product C','Product D']].plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Stacked Monthly Sales')
plt.ylabel('Sales')
plt.show()

# Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df[['Product A','Product B','Product C','Product D']], annot=True, cmap='YlGnBu', xticklabels=True)
plt.title('Monthly Sales Heatmap')
plt.show()

# Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(data=df[['Product A','Product B','Product C','Product D']])
plt.title('Sales Distribution per Product')
plt.show()
```
